```python
%matplotlib notebook
%config InlineBackend.print_figure_kwargs={'bbox_inches':None}

n_animals = 10
n_steps = 100
diffusion_speed = 0.005   # Something speed-like

x = np.ones(n_animals)*0.5
y = np.ones(n_animals)*0.5

vx = np.random.rand(n_animals)
vy = np.random.rand(n_animals)

# Plotting
fig = plt.figure(figsize=(3,3))
ax = plt.subplot(111)
plt.xlim((0,1));
plt.ylim((0,1));
fig.show()
fig.canvas.draw()

for t in range(n_steps):
    x += np.random.randn(n_animals)*diffusion_speed
    y += np.random.randn(n_animals)*diffusion_speed
    for a in range(n_animals):
        x = x + vx * diffusion_speed
        y = y + vy * diffusion_speed
        
        x = np.where(x>0, x, vx)
        x = np.where(x<1, x, vx)
        y = np.where(y>0, y, vy)
        y = np.where(y<1, y, vy)

    
    ax.clear()
    plt.plot(x,y,'k.',markersize=2);
    plt.xlim((0,1));
    plt.ylim((0,1));
    fig.canvas.draw() # <--------------- The main command to make it move
    #time.sleep(0.01)
```
